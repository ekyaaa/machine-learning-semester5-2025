```python
dpath = '/content/drive/MyDrive/ML_Dataset/tg3/wbc.csv'
df = pd.read_csv(dpath)

# 1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
df_clean = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')
df_clean.head()
```

```python
# 2. Lakukan proses encoding pada kolom "diagnosis".
df['diagnosis'].head()
df['diagnosis'].describe()


le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
df.head()
```

```python
df['diagnosis'].describe()
```

```python
# 3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

# Gunakan dataframe yang sudah dibersihkan
df_clean = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')

# Pisahkan target dan fitur
y = df_clean['diagnosis']
X = df_clean.drop('diagnosis', axis=1)

# Ambil hanya kolom numerik
numeric_cols = X.select_dtypes(include=["number"]).columns

constant_cols = [col for col in numeric_cols if X[col].std() == 0]
print("Kolom konstan:", constant_cols)

X = X.drop(columns=constant_cols)

# Standardisasi
scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# Gabungkan kembali dengan target
df_scaled = pd.concat([y, X], axis=1)

df_scaled.head()
```

```python
# 4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.
# Pisahkan fitur dan target
X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]

# Stratified split dengan rasio 80:20
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("Jumlah data latih:", X_train.shape[0])
print("Jumlah data uji  :", X_test.shape[0])
print("Distribusi y_train:", y_train.value_counts(normalize=True))
print("Distribusi y_test :", y_test.value_counts(normalize=True))
```
